Introduction to Primality Test and School Method

Given a positive integer, check if the number is prime or not. A prime is a natural number greater than 1 that has no positive divisors other than 1 and itself. Examples of the first few prime numbers are {2, 3, 5, ...}

Examples : 
Input:  n = 11
Output: true

Input:  n = 15
Output: false

Input:  n = 1
Output: false 

School Method

A simple solution is to iterate through all numbers from 2 to n-1 and for every number check if it divides n. If we find any number that divides, we return false. 

````python
def isPrime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

if __name__ == "__main__":
    n = int(input("Enter a number: "))
    print(f"Input: n = {n}, and is it prime? {isPrime(n)}")
````

Optimized School Method

We can do the following optimizations: Instead of checking till n, we can check till ‚àön because a larger factor of n must be a multiple of a smaller factor that has been already checked. The implementation of this method is as follows:

````python
import math

def isPrime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

if __name__ == "__main__":
    n = int(input("Enter a number: "))
    print(f"Input: n = {n}, and is it prime? {isPrime(n)}")
````

## [Approach 3] 6k ¬± 1 Optimisation ‚Äî O(‚àön / 3) Time and O(1) Space

This approach cuts the work to roughly **‚Öì of the ‚àön method** by exploiting a number theory property.

---

### Key Observation

Every integer can be written as one of:

```
6k + 0,  6k + 1,  6k + 2,  6k + 3,  6k + 4,  6k + 5
```

Of these six forms, **four are always composite**:
| Form | Always divisible by | So‚Ä¶ |
|------|---------------------|-----|
| 6k + 0 | 2 and 3 | composite |
| 6k + 2 | 2 | composite |
| 6k + 3 | 3 | composite |
| 6k + 4 | 2 | composite |

That leaves only **6k + 1** and **6k + 5 (= 6k ‚àí 1)** as *candidates* for primes greater than 3.

> So instead of checking every number up to ‚àön, we only need to check numbers of the form **6k ¬± 1**.

---

### Algorithm ‚Äî Step by Step

**Step 1 ‚Äî Handle small cases directly**
- `n ‚â§ 1` ‚Üí not prime
- `n == 2` or `n == 3` ‚Üí prime (special-case before the division checks)

**Step 2 ‚Äî Eliminate obvious composites in O(1)**
- If `n % 2 == 0` ‚Üí divisible by 2, not prime
- If `n % 3 == 0` ‚Üí divisible by 3, not prime

**Step 3 ‚Äî Loop over 6k ¬± 1 candidates only**
- Start at `i = 5` (first candidate: 6¬∑1 ‚àí 1)
- Each step jumps by **6** (to the next pair `6k‚àí1, 6k+1`)
- Check both `i` (= 6k ‚àí 1) and `i + 2` (= 6k + 1)
- Stop when `i > ‚àön`

```
i = 5  ‚Üí check 5 and 7
i = 11 ‚Üí check 11 and 13
i = 17 ‚Üí check 17 and 19
‚Ä¶
```

---

### Why is this ~3√ó faster than the ‚àön method?

| Method | Candidates checked up to ‚àön |
|---|---|
| Optimised ‚àön (Approach 2) | Every integer ‚Üí **‚àön** checks |
| **6k ¬± 1 (Approach 3)** | Only 2 out of every 6 ‚Üí **‚àön / 3** checks |

The constant factor improves by ~3√ó, but the **asymptotic class remains O(‚àön)** ‚Äî only the leading constant shrinks.

````python
def isPrime(n):
    if n == 2 or n == 3:
        return True
    if n <= 1 or n % 2 == 0 or n % 3 == 0:
        return False
    
    # To check through all numbers of the form 6k ¬± 1
    # until i <= square root of n, with step value 6
    for i in range(5, int(math.sqrt(n)) + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True

if __name__ == "__main__":
    n = int(input("Enter a number: "))
    print(f"Input: n = {n}, and is it prime? {isPrime(n)}")
````
