Write a program to reverse digits of a number

Given an Integer n, find the reverse of its digits.

Examples:  

Input: n = 122
Output: 221
Explanation: By reversing the digits of number, number will change into 221.

Input: n = 200
Output: 2
Explanation: By reversing the digits of number, number will change into 2.

Input: n = 12345 
Output: 54321
Explanation: By reversing the digits of number, number will change into 54321.

Reversing Digit by Digit

This idea for this approach is to repeatedly extract the last digit of n using the modulus operator (n % 10) and appending it to the reverse number (revNum). After extracting the digit, the number n is reduced by dividing it by 10 (n = n / 10). This process continues until n becomes 0. Finally, the reversed number (revNum) is returned.

```python
def reverseDigits(n):
    revNum = 0
    while n > 0:
        revNum = revNum * 10 + n % 10
        n = n // 10
    return revNum

if __name__ == "__main__":
    n = int(input("Enter a number: "))
    print(f"Input: n = {n}, and the reversed number is: {reverseDigits(n)}")
```

## Using Recursion (Call Stack Unwind + basePos)

This approach recurses **all the way to the leftmost digit first**, then does the work **on the way back up** (unwinding). The call stack itself acts as the reversal ‚Äî leftmost digit is processed first at `basePos=1`, and as the stack unwinds, `basePos` grows for the rightmost digits.

**Trace with `n = 4562` (expected `2654`):**

| Unwind step | digit processed | basePos | contribution | revNum |
|-------------|----------------|---------|-------------|--------|
| 1 (deepest) | 4 | 1 | 4 √ó 1 = 4 | 4 |
| 2 | 5 | 10 | 5 √ó 10 = 50 | 54 |
| 3 | 6 | 100 | 6 √ó 100 = 600 | 654 |
| 4 (top) | 2 | 1000 | 2 √ó 1000 = 2000 | **2654** ‚úì |


```python
def reverseDigits(n, revNum, basePos):
    if n > 0:
        reverseDigits(n // 10, revNum, basePos)   # recurse first (go deep)
        revNum[0] += (n % 10) * basePos[0]        # work on the way back up
        basePos[0] *= 10

if __name__ == "__main__":
    n = int(input("Enter a number: "))
    revNum = [0]
    basePos = [1]
    reverseDigits(n, revNum, basePos)
    print(f"Input: n = {n}, and the reversed number is: {revNum[0]}")
```

Using String and Slicing in Python

The approach used is "Using Slicing". This technique involves converting the number into a string, then reversing that string by using slicing operations. After reversing, the string is converted back into a number. This method is simple and efficient but requires additional space for storing the string representation of the number.

```python
def reverseDigits(n):
    # Convert the number to a string, reverse it using slicing, and convert back to an integer
    return int(str(n)[::-1])

if __name__ == "__main__":
    n = int(input("Enter a number: "))
    print(f"Input: n = {n}, and the reversed number is: {reverseDigits(n)}")
```
