Find Closest to n and Divisible by m

Given two integers n and m (m != 0). Find the number closest to n and divisible by m. If there is more than one such number, then output the one having maximum absolute value.

Examples: 

Input: n = 13, m = 4
Output: 12
Explanation: 12 is the closest to 13, divisible by 4.

Input: n = -15, m = 6
Output: -18
Explanation: Both -12 and -18 are closest to -15, but -18 has the maximum absolute value.

[Naive Approach] Iterative Checking - O(m) Time and O(1) Space

The basic idea is to start checking from n - m to n + m one by one and take the closest number.

```python
def closest_number(n, m):
    # find the quotient
    closest = 0
    min_difference = float('inf')

    # Check numbers around n
    for i in range(n - abs(m), n + abs(m) + 1):
        if i % m == 0:
            difference = abs(n - i)

            if difference < min_difference or \
            			(difference == min_difference and abs(i) > abs(closest)):
                closest = i
                min_difference = difference
    return closest

if __name__ == "__main__":
    n = int(input("Enter a number (n): "))
    m = int(input("Enter a number (m): "))
    result = closest_number(n, m)
    print(f"The closest number to {n} that is divisible by {m} is: {result}")
```

[Expected Approach] By finding Quotient - O(1) Time and O(1) Space

Goal

Find integer k √ó m that is closest to n
(if two have same distance ‚Üí pick the one with larger |value|)

Core mathematical steps

1. Take m_positive = |m|
    (we always work with positive divisor ‚Äî sign of m doesn't change the multiples)
2. Compute quotient
    q = ‚åä n / m_positive ‚åã
    (floor = greatest integer ‚â§ n / m_positive)
3. The two mathematically closest multiples are almost always these two:
    A = q √ó m_positive √ó sign(m)
    B = (q + 1) √ó m_positive √ó sign(m)
    ‚Üí these are the two multiples of m that sandwich n (or touch it)
4. Compute distances
    dist_A = |n ‚àí A|
    dist_B = |n ‚àí B|
5. Decision rule (this is the complete logic):
    if dist_A < dist_B  ‚Üí answer = A
    if dist_B < dist_A  ‚Üí answer = B
    if dist_A = dist_B  ‚Üí answer = whichever of A or B has larger |x|


Worked examples step by step (pure math)

n = 13, m = 4
m_positive = 4
q = ‚åä13 / 4‚åã = ‚åä3.25‚åã = 3
A = 3 √ó 4 = 12
B = 4 √ó 4 = 16
dist_A = |13 ‚àí 12| = 1
dist_B = |13 ‚àí 16| = 3
1 < 3 ‚Üí answer = 12

```python
def closestNumber(n, m):
    print("=" * 40)
    print(f"  Input: n = {n}, m = {m}")
    print("=" * 40)

    if m == 0:
        print("  m = 0 ‚Üí returning n as-is")
        return n

    # Step 1
    m_positive = abs(m)
    print(f"\nStep 1 ‚îÇ m_positive = |m| = |{m}| = {m_positive}")

    # Step 2
    q = n // m_positive
    print(f"Step 2 ‚îÇ q = ‚åän / m_positive‚åã = ‚åä{n} / {m_positive}‚åã = {q}")

    # Step 3
    A = q * m_positive
    B = (q + 1) * m_positive
    print(f"Step 3 ‚îÇ Two candidate multiples that bracket n:")
    print(f"       ‚îÇ   A = q √ó m_positive         = {q} √ó {m_positive} = {A}")
    print(f"       ‚îÇ   B = (q + 1) √ó m_positive   = {q + 1} √ó {m_positive} = {B}")

    # Step 4
    dist_A = abs(n - A)
    dist_B = abs(n - B)
    print(f"Step 4 ‚îÇ Distances from n = {n}:")
    print(f"       ‚îÇ   dist_A = |{n} ‚àí {A}| = {dist_A}")
    print(f"       ‚îÇ   dist_B = |{n} ‚àí {B}| = {dist_B}")

    # Step 5
    print(f"Step 5 ‚îÇ Decision:")
    if dist_A < dist_B:
        print(f"       ‚îÇ   dist_A ({dist_A}) < dist_B ({dist_B}) ‚Üí answer = A = {A}")
        result = A
    elif dist_B < dist_A:
        print(f"       ‚îÇ   dist_B ({dist_B}) < dist_A ({dist_A}) ‚Üí answer = B = {B}")
        result = B
    else:
        chosen = A if abs(A) >= abs(B) else B
        print(f"       ‚îÇ   Tie! dist_A = dist_B = {dist_A}")
        print(f"       ‚îÇ   |A| = |{A}| = {abs(A)},  |B| = |{B}| = {abs(B)}")
        print(f"       ‚îÇ   Pick larger |value| ‚Üí answer = {chosen}")
        result = chosen

    print(f"\n  ‚úì Result: {result}")
    print("=" * 40)
    return result

if __name__ == "__main__":
    n = int(input("Enter a number (n): "))
    m = int(input("Enter a number (m): "))
    closestNumber(n, m)
```

```python
def closestNumber(n, m):
    if m == 0:
        return n

    m_positive = abs(m)
    q = n // m_positive

    A = q * m_positive
    B = (q + 1) * m_positive

    dist_A = abs(n - A)
    dist_B = abs(n - B)

    if dist_A < dist_B:
        return A
    elif dist_B < dist_A:
        return B
    else:
        return A if abs(A) >= abs(B) else B

if __name__ == "__main__":
    n = int(input("Enter a number (n): "))
    m = int(input("Enter a number (m): "))
    result = closestNumber(n, m)
    print(f"The closest number to {n} that is divisible by {m} is: {result}")
```
